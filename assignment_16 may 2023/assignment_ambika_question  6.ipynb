#### since data type of next evolution and previous evolution are of string as well as float( dtype of nan is float), we need to convert dtype to string so that we can split it to make a list, then  the length of list returns number of evolutions

```python
df['next_evolution'] = df['next_evolution'].astype('str')
df['prev_evolution'] = df['prev_evolution'].astype('str')
```

```python
for i in range(151):
    df['next_evolution'][i] = df['next_evolution'][i].split('},')
    
for i in range(151):
    df['prev_evolution'][i] = df['prev_evolution'][i].split('},')    
```

```python
df['next_evl_count'] = df['next_evolution'].apply(lambda x:0 if x==['nan'] else len(x))
df['prev_evl_count'] = df['prev_evolution'].apply(lambda x:0 if x==['nan'] else len(x))
```

```python
current_life = 1
df['total_evolutions'] = df['prev_evl_count'] + current_life + df['next_evl_count']
```

### Pokemons that do not have more than 2 evolutions

```python
df[df['total_evolutions'] <= 2][['name','prev_evolution','next_evolution','total_evolutions']] 
```

**Get all Pokemons whose spawn time is less than 300 seconds**

```python
df.head()
```

```python
type(df['spawn_time'][0])
```

```python
df['spawn_time'] = df['spawn_time'].astype('str')
```

```python
for i in range(151):
    df['spawn_time'][i] = df['spawn_time'][i].split(':')
```

```python
df['spawn_min'] = df['spawn_time'].apply(lambda x:0 if x==['nan'] else x[0]).astype('int')
df['spawn_sec'] = df['spawn_time'].apply(lambda x:0 if x==['nan'] else x[1]).astype('int')
```

```python
df['spawn_time_seconds'] = (df['spawn_min']*60) + df['spawn_sec']
```

### Pokemons whose spawn time is less than 300 seconds

```python
df[df['spawn_time_seconds']<300][['name','spawn_time','spawn_time_seconds']]
```
