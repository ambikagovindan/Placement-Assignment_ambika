**Get all the overall ratings for each season and using plots compare the ratings for all the
seasons, like season 1 ratings, season 2, and so on.**

## average ratings for all seasons

```python
plt.figure(figsize=(15,15))
sns.set_style('whitegrid')
sns.lineplot(x='airdate',y='average_rating',data = df,hue='season',palette='bright')
plt.xticks(rotation = 90)
plt.title('Average ratings on different seasons')
```

**Get all the episode names, whose average rating is more than 8 for every season**

## episodes whose average rating is more than 8 for every season

```python
df[df['average_rating']>8][['name','number','average_rating','season']]
```

**Get all the episode names that aired before May 2019**

## episodes  that aired before May 2019

```python
df[df['airdate']< '2019-05-01']
```

## Get the episode name from each season with the highest and lowest rating

```python
df1 = df[df['season']==1]
df2 = df[df['season']==2]
df3 = df[df['season']==3]
df4 = df[df['season']==4]
```

```python
print('Season:1')
print('Minimum:rating:',min(df1.average_rating))
print(df1[df1.average_rating==df1.average_rating.min()][['name','average_rating']])
print('-'*30)
print('Maximum rating:',max(df1.average_rating))
print(df1[df1.average_rating==df1.average_rating.max()][['name','average_rating']])
print('*'*50)

print('Season:2')
print('Minimum:rating:',min(df2.average_rating))
print(df2[df2.average_rating==df2.average_rating.min()][['name','average_rating']])
print('-'*30)
print('Maximum rating:',max(df2.average_rating))
print(df2[df2.average_rating==df2.average_rating.max()][['name','average_rating']])
print('*'*50)

print('Season:3')
print('Minimum:rating:',min(df3.average_rating))
print(df3[df3.average_rating==df3.average_rating.min()][['name','average_rating']])
print('-'*30)
print('Maximum rating:',max(df3.average_rating))
print(df3[df3.average_rating==df3.average_rating.max()][['name','average_rating']])
print('*'*50)


print('Season:4')
print('Minimum:rating:',min(df4.average_rating))
print(df4[df4.average_rating==df4.average_rating.min()][['name','average_rating']])
print('-'*30)
print('Maximum rating:',max(df4.average_rating))
print(df4[df4.average_rating==df4.average_rating.max()][['name','average_rating']])
print('*'*50)
```

## Get the summary for the most popular ( ratings ) episode in every season

```python
pd.set_option('display.max_colwidth', None)
print(df1[df1.average_rating==df1.average_rating.max()][['season','name','summary']])
print('-'*70)

print(df2[df2.average_rating==df2.average_rating.max()][['season','name','summary']])
print('-'*70)

print(df3[df3.average_rating==df3.average_rating.max()][['season','name','summary']])
print('-'*70)

print(df4[df4.average_rating==df4.average_rating.max()][['season','name','summary']])
print('-'*70)
```
