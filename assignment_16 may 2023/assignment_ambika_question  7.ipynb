**Get all the Earth meteorites that fell before the year 2000**

```python
# checking for null value
df[df['year'].isnull()]
```

```python
# dropping null value raw
df1 = df.drop(df.index[37])
```

```python
df1['year'] = df1['year'].apply(lambda x:x[0:4]).astype('int')
```

### Earth meteorites that fell before the year 2000

```python
pd.options.display.max_rows=999
df1[df1['year'] < 2000][['name','year']]
```

**Get all the earth meteorites co-ordinates who fell before the year 1970**

```python
# feature enginering
df1['geolocation'] = df1['geolocation'].astype('str')
df1['geolocation'] = df1['geolocation'].apply(lambda x:'nan' if x=='nan' else x.split("'coordinates':")[1].strip('}'))
```

 ### earth meteorites co-ordinates who fell before the year 1970

```python
df1[df1['year'] < 1970][['name','year','geolocation']]
```

**Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more

### earth meteorites with mass more than 10000kg

```python
df[df['mass'] > 10000][['name','mass']]
```
