```python
# downloading the json data from the link

import urllib.request
import json 
your_url = 'https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'
with urllib.request.urlopen(your_url) as url:
    data = json.loads(url.read().decode())
data
```

```python
type(data)
```

```python
data.keys()
```

```python
data_1 = data.values()
data_1
```

```python
type(data_1)
```

```python
# converting dictionary values to a list

data_2 = list(data_1)
data_2
```

```python
len(data_2)
```

```python
data_2[0][0]
```

```python
len(data_2[0])
```

```python
#creating a 1 dimensional list of data

data3 = []

for i in range (0,len(data_2[0])):
    data3.append(data_2[0][i])
    
data3
```

```python
# creating pandas dataframe

final_data = pd.DataFrame(data3)
final_data
```

```python
final_data.to_excel('Pokeman_data.xlsx',index=False)
```
